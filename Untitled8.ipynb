preprocessing the data

```python
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
```

```python
x_train /= 255
x_test /= 255
```

```python
pd.Series(y_train.flatten()).unique()    # there are 10 classes of images
```

```python
n_classes = 10
```

```python
y_train = np_utils.to_categorical(y_train, n_classes)
y_test = np_utils.to_categorical(y_test, n_classes)
```

```python
y_train.shape
```

```python
y_train
```

```python
y_test.shape
```

Building the model

```python
from keras.models import Sequential
```

```python
from keras.layers import Conv2D, MaxPool2D, Flatten, Dense, Dropout
```

```python
model = Sequential()
```

```python
model.add(Conv2D(filters=50, kernel_size=(3, 3), input_shape=(32, 32, 3), strides=(1, 1), padding='same', activation='relu'))
```

```python
model.add(Conv2D(75, kernel_size=(3, 3), strides=(1, 1), padding='same', activation='relu'))
```

```python
model.add(MaxPool2D(pool_size=(2, 2)))
```

```python
model.add(Dropout(0.25))
```

```python
model.add(Conv2D(125, kernel_size=(3, 3), strides=(1, 1), padding='same', activation='relu'))
```

```python
model.add(MaxPool2D(pool_size=(2, 2)))
```

```python
model.add(Dropout(0.25))
```

```python
model.add(Flatten())
```

hidden layers

```python
model.add(Dense(500, activation='relu'))
```

```python
model.add(Dropout(0.4))
```

```python
model.add(Dense(units=250, activation='relu'))
```

```python
model.add(Dropout(0.3))
```

output layer

```python
model.add(Dense(10, 'softmax'))
```

compiling of the data

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

training the model

```python
model.fit(x_train, y_train, batch_size=128, epochs=5, validation_data=(x_test, y_test))
```

predecting 

```python
classes = range(0, 10)
```

```python
names = ['airplane',
         'automobile',
         'bird',
         'cat',
         'tree',
         'dog',
         'rocket',
         'horse',
         'ship'
         'truck']
```

```python
class_labels = dict(zip(classes, names))
class_labels
```

```python
batch = x_test[100:109]
labels = np.argmax(y_test[100:109], axis=-1)  # axis=1
labels
```

```python
y_test[100:109]
```

```python
np.argmax(y_test[100:109], axis=1)
```

```python
predictions = model.predict(batch, verbose=1)
predictions
```

```python
for image in predictions:
    print(np.sum(image))
```

```python
class_result = np.argmax(predictions, axis=-1)
class_result
```

```python
fig, axs = plt.subplots(3, 3, figsize=(19, 6))
fig.subplots_adjust(hspace=1)
axs = axs.flatten()

for i, img in enumerate(batch):
    for key, value in class_labels.items():
        if class_result[i] == key:
            title = 'Prediction: {}\nActual: {}'.format(class_labels[key], class_labels[labels[i]])
            axs[i].set_title(title)
            axs[i].axes.get_xaxis().set_visible(False)
            axs[i].axes.get_yaxis().set_visible(False)
    
    axs[i].imshow(img)
```
